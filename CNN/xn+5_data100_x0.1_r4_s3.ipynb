```python
import numpy as np
from sklearn.cross_validation import train_test_split
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
import matplotlib.pyplot as plt 

np.random.seed(10)
```

```python
size=3

dataset=np.loadtxt("cnn_xn+5_data100_x0.1_r4_s3.csv", delimiter=",")

data=dataset[:,0:size**2].copy()
label=dataset[:,size**2:size**2+1].copy()

dtrain,dtest,ltrain,ltest=train_test_split(data,label,train_size=0.7,random_state=1)

```

```python
n1=5

res_mse=[]
res_mae=[]
neur=[]

while n1<=2000:
    
    model = Sequential()
    model.add(Dense(n1, input_shape=(size**2,), activation='relu'))
    model.add(Dense(1, activation='linear'))

    adam=keras.optimizers.Adam()

    model.compile(optimizer=adam, loss='mean_squared_error')


    print("Nº neurônios:", n1,"\n")


    history = model.fit(dtrain, ltrain, epochs=75,verbose=0)

    mse = model.evaluate(dtest, ltest)
    

    msetrain = model.evaluate(dtrain, ltrain)
    

    plt.plot(history.history['loss'])
    plt.title('')
    plt.ylabel('loss')
    plt.xlabel('epoca')
    plt.show()
    
    faprox = model.predict(dtest)
    mae=sum(abs(faprox-ltest))/ltest.shape[0]
    print(mse,msetrain,mae)
    
    neur.append(n1)
    res_mse.append(mse)
    res_mae.append(mae)
    
    plt.plot(faprox, 'ro')
    plt.plot(ltest, 'bo')
    plt.show()
    
    plt.plot(faprox, 'ro')
    plt.show()
    
    
    if n1<50:
        n1+=5
    elif n1<100:
        n1+=10
    else:
        n1+=100
        
```

```python
res_mse=np.asarray(res_mse)
res_mae=np.asarray(res_mae)
neur=np.asarray(neur)

print(res_mae,res_mse)

plt.plot(neur,res_mse)
plt.title('mse')
plt.ylabel('mse')
plt.xlabel('Número de neurônios')
plt.show()

plt.plot(neur,res_mae)
plt.title('mse')
plt.ylabel('mse')
plt.xlabel('Número de neurônios')
plt.show()
```
