```python
import numpy as np
```

```python
r = 3.8
x0=0.1

iters=1000

ini=1
f_range=20

file = open('data1000_1-20_r3.8.csv','w+')
```

```python

for i in range(200): 
    x1=x0*r*(1-x0)
    x1 = round(x1,15) 
    x0=x1
    


x_o=[]
for i in range(iters+f_range):
    x1=x0*r*(1-x0)
    x1 = round(x1,15)
    x0=x1
    x_o.append(x1)

x_o=np.asarray(x_o)

for i in range (iters):
    for j in range (ini,f_range+1):
        print(j)
        
        file.write('{}'.format(x_o[i]))
        file.write(',')
        file.write('{}'.format(j))
        file.write(',')
        file.write('{}'.format(x_o[i+j]))
        file.write('\n')
    
        
file.close()
```
