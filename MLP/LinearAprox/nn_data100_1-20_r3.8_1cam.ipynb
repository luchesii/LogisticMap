```python
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
import matplotlib.pyplot as plt 
from sklearn.cross_validation import train_test_split
from sklearn.metrics import confusion_matrix
import itertools

np.random.seed(10)

data=np.loadtxt("data100_1-20_r3.8.csv", delimiter=",")


dataset=data[:,0:2].copy()
label=data[:,2:3].copy()
    
print(label)
dtrain,dtest,ltrain,ltest=train_test_split(dataset,label,train_size=0.7,random_state=10)

```

```python
file= open("results_data100_1-20_r3.8_tr.csv","w+")

n1=5


while n1<=1000:
    
    model = Sequential()
    model.add(Dense(n1, input_shape=(2,), activation='relu'))
    model.add(Dense(1, activation='linear'))

    #model.summary()

    adam=keras.optimizers.Adam()

    model.compile(optimizer=adam, loss='mean_squared_error')


    print("Nº neurônios:", n1,"\n")


    history = model.fit(dtrain, ltrain, epochs=300, batch_size=10,verbose=0)

    mse = model.evaluate(dtest, ltest)
    

    msetrain = model.evaluate(dtrain, ltrain)
    

    file.write('{}'.format(n1))
    file.write(',')
    file.write('{}'.format(msetrain))
    file.write(',')
    file.write('{}'.format(mse))
    file.write('\n')
    

    plt.plot(history.history['loss'])
    plt.title('Aprox 100, futuro1-20, r3.8 camada oculta neur=%i' %n1)
    plt.ylabel('loss')
    plt.xlabel('epoca')
    plt.show()
    
    faprox = model.predict(dtest)
    
    plt.plot(faprox, 'ro')
    plt.plot(ltest, 'bo')
    plt.show()
    
    
    data_t1=np.loadtxt("test.csv", delimiter=",")

    t1=data_t1[:,0:2].copy()
    l1=data_t1[:,2:3].copy()
    
    f_t1 = model.predict(t1)
    
    plt.plot(f_t1, 'ro')
    plt.plot(l1, 'bo')
    plt.title('t1')
    plt.show()
    
    data_t2=np.loadtxt("test2.csv", delimiter=",")

    t2=data_t2[:,0:2].copy()
    l2=data_t2[:,2:3].copy()
    
    f_t2 = model.predict(t2)
    
    plt.plot(f_t2, 'ro')
    plt.plot(l2, 'bo')
    plt.title('t2')
    plt.show()
    
    for i in range(ltest.shape[0]): print(faprox[i],ltest[i],np.abs(faprox[i]-ltest[i]))
    

    if n1<50:
        n1+=5
    elif n1<100:
        n1+=10
    else:
        n1+=100
        
        

    
```

```python
file.close()
```
