```python
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
from sklearn.cross_validation import train_test_split



dataset=np.loadtxt("dz-logis_data1000_d10_v3.csv", delimiter=",")
label=np.loadtxt("dz-logis_lab_data1000_d10_v3.csv", delimiter=",")

dtrain,dtest,ltrain,ltest=train_test_split(dataset,label,train_size=0.7,random_state=1)

file= open("dz-results_logis_data1000_d10_v3-cop.csv","w+")

def one_hot_encode_object_array(arr):
    '''One hot encode a numpy array of objects (e.g. strings)'''
    uniques, ids = np.unique(arr, return_inverse=True)
    return np_utils.to_categorical(ids, len(uniques))
ltrain=one_hot_encode_object_array(ltrain)
ltest=one_hot_encode_object_array(ltest)
```

```python
n1=1

file.write('adam\n')

while n1<=40:
    
    model = Sequential()
    model.add(Dense(n1, input_shape=(3,), activation='relu'))
    model.add(Dense(10, activation='softmax'))

    model.summary()

    adam=keras.optimizers.Adam()

    model.compile(optimizer=adam, loss='categorical_crossentropy', metrics=['accuracy'])


    print("1 hidden layer:", n1,"\n")


    model.fit(dtrain, ltrain, epochs=300, batch_size=10)

    loss,accur = model.evaluate(dtest, ltest)

    losstrain,acctrain = model.evaluate(dtrain, ltrain)
    
    acctrain=round(acctrain*100,2)
    accur=round(accur*100,2)

    print(loss, accur)
    file.write('{}'.format(n1))
    file.write(',')
    file.write('{}'.format(acctrain))
    file.write(',')
    file.write('{}'.format(accur))
    file.write('\n')
        
    n1+=1

    
```
