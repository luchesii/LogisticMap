```python
import numpy as np
import operator

b=np.loadtxt("logis_lab_data1000_d100_v2.csv", delimiter=",")

i=ct=0
contador = {} #armazena quandos votos tem cada categoria

for i in range(b.shape[0]):
    a=b[i]
    
    if a in contador:
        contador[a] += 1
    else:
        contador[a] = 1
        ct+=1

listavotos = sorted(contador.items(), key=operator.itemgetter(1), reverse=True)
lista=np.asarray(listavotos)
print(lista.shape,lista)
```
