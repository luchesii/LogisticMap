```python
import numpy as np
import random

np.random.seed(19)

iter=10000
n_elem=2

x0=0.1
r=3.8

q=2
prec=15

x=np.empty((n_elem*iter,1),dtype=np.double)
y=np.empty((n_elem,1),dtype=np.double)

file= open("logis_data10000_d100_v2.csv","w+")
file2=open("logis_lab_data10000_d100_v2.csv", "w+")
```

```python

for i in range (200): #deve ser >3.3*d (onde d são digitos de precisão)
    x1=r*x0*(1-x0)
    x1=round(x1,prec)
    x0=x1
```

```python
x[0]=x0
for i in range (n_elem*iter-1):
    x1=r*x0*(1-x0)
    x1=round(x1,prec)
    x[i+1]=x1
    x0=x1
print (x)
```

```python
ind=random.sample(range((n_elem*iter)-n_elem-1), iter)


for i in range (iter):
    for j in range (n_elem):
        y[j]=x[ind[i]+j]
        
    yp=float(x[ind[i]+n_elem])
    
    y_l=y.tolist()
    y_s=','.join(map(str, y_l))
    y_s=y_s.replace('[','').replace(']','')

    file.write(y_s)
    file.write("\n")
    
    s=str(yp*pow(10,q))[0:q]
    print(y,yp,s)
    
    file2.write(s)
    file2.write('\n')
        
    
    

```
