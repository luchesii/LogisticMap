```python
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
import matplotlib.pyplot as plt 
from sklearn.cross_validation import train_test_split
from sklearn.metrics import confusion_matrix
import itertools

np.random.seed(10)

dataset=np.loadtxt("3logis_data10000_d10_v2.csv", delimiter=",")
label=np.loadtxt("3logis_lab_data10000_d10_v2.csv", delimiter=",")

#for i in range(label.shape[0]):
    #label[i]=label[i]-1
    
dtrain,dtest,ltrain,ltest=train_test_split(dataset,label,train_size=0.7,random_state=1)

ltest2=ltest
file= open("3results_logis_data10000_d10_v2.csv","w+")
```

```python
#função tirada do sklearn

def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()

```

```python
n1=5


while n1<=1000:
    
    model = Sequential()
    model.add(Dense(n1, input_shape=(2,), activation='relu'))
    model.add(Dense(10, activation='softmax'))

    #model.summary()

    adam=keras.optimizers.Adam()

    model.compile(optimizer=adam, loss='sparse_categorical_crossentropy', metrics=['accuracy'])


    print("Nº neurônios:", n1,"\n")


    history = model.fit(dtrain, ltrain, epochs=300, batch_size=10,verbose=0)

    loss,accur = model.evaluate(dtest, ltest)

    losstrain,acctrain = model.evaluate(dtrain, ltrain)
    
    acctrain=round(acctrain*100,2)
    accur=round(accur*100,2)

    print(loss, accur)
    file.write('{}'.format(n1))
    file.write(',')
    file.write('{}'.format(acctrain))
    file.write(',')
    file.write('{}'.format(accur))
    file.write('\n')
    
    plt.plot(history.history['acc'])
    plt.title('Xn+3, Data:10000, div=10, v=2, neur=%i' %n1)
    plt.ylabel('Acurácia')
    plt.xlabel('epoca')
    plt.show()
    
    plt.plot(history.history['loss'])
    plt.title('Xn+3, Data:10000, div=10, v=2, neur=%i' %n1)
    plt.ylabel('loss')
    plt.xlabel('epoca')
    plt.show()
    
    MCon = confusion_matrix(ltest2,model.predict_classes(dtest,batch_size=10,verbose=0))
    
    Mcon_lab=['1','2','3','4','5','6','7','8','9']
    
    plot_confusion_matrix(MCon,Mcon_lab,'Matriz de confusão')
    
    plt.show()
    
    if n1<50:
        n1+=5
    elif n1<100:
        n1+=10
    else:
        n1+=100
        
        

    
```
