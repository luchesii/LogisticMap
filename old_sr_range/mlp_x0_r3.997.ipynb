```python
import numpy as np
from sklearn.model_selection import train_test_split
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
import matplotlib.pyplot as plt 

np.random.seed(10)

def get_next(x0,fut):
    r = 3.997
    for i in range(fut):
        x1=x0*r*(1-x0)
        x0=x1
    return(x1)

```

```python
x0 = np.loadtxt("esn_random_x0.1_r3.997.csv", delimiter=" ")

xn = np.empty((x0.shape[0]))

for i in range(x0.shape[0]):
    xn[i] = get_next(x0[i],1)
    
x0train,x0test,xntrain,xntest=train_test_split(x0,xn,train_size=0.7,random_state=1)
```

```python
n=5

MSE = []
MAE = []
N = []

while n<=2000:
    
    model = Sequential()
    model.add(Dense(n, input_shape=(1,), activation='relu'))
    model.add(Dense(1, activation='linear'))

    adam=keras.optimizers.Adam()

    model.compile(optimizer=adam, loss='mean_squared_error')


    print("Nº neurônios:", n,"\n")


    history = model.fit(x0train, xntrain, epochs=100,verbose=0)

    mse = model.evaluate(x0test, xntest)
    

    msetrain = model.evaluate(x0train, xntrain)
    

    plt.plot(history.history['loss'])
    plt.title('')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.show()
    
    faprox = model.predict(x0test).T
    mae = np.sum(np.abs(faprox-xntest))/xntest.shape[0]
    print(mse,msetrain,mae)
    
    N.append(n)
    MSE.append(mse)
    MAE.append(mae)
    
    
    if n < 50:
        n+=5
    elif n < 100:
        n+=10
    else:
        n+=100
```

```python
RMSE=np.sqrt(MSE)

ind = np.argsort(MSE)
for i in range(len(MSE)): 
    ii = ind[i]
    print(N[ii],MSE[ii])
    
ind = np.argsort(MAE)
for i in range(len(MAE)): 
    ii = ind[i]
    print(N[ii],MAE[ii])
    
ind = np.argsort(RMSE)
for i in range(len(RMSE)): 
    ii = ind[i]
    print(N[ii],RMSE[ii])
    
plt.plot(N,MSE)
plt.title('MSE xn+1')
plt.ylabel('mse')
plt.xlabel('Number of neurons')
plt.show()

plt.plot(N,RMSE)
plt.title('RMSE xn+1')
plt.ylabel('rmse')
plt.xlabel('Number of neurons')
plt.show()



plt.plot(N,MAE)
plt.title('MAE xn+1')
plt.ylabel('mae')
plt.xlabel('Number of neurons')
plt.show()

print('MSE min:',N[np.argmin(MSE)],MSE[np.argmin(MSE)])
print('MAE min:',N[np.argmin(MAE)],MAE[np.argmin(MAE)])
print('RMSE min:',N[np.argmin(RMSE)],RMSE[np.argmin(RMSE)])
```
