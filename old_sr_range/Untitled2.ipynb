```python
import numpy as np

rmse = np.loadtxt("esn_big_minstable_r4_x0.1_d1000.csv", delimiter=",")

rmse = rmse[:,-1]
rmse_2d = []
rmse_3d = []
rmse_4d = []

for i in range (rmse.shape[0]):
    if rmse[i]>=0.01:
        rmse_2d.append(rmse[i])
        print(rmse[i])
    elif rmse[i]>=0.001:
        rmse_3d.append(rmse[i])
    elif rmse[i]>=0.0001:
        rmse_4d.append(rmse[i])
    else:
        print(rmse[i])

#rint('min',np.amin(rmse_2d),'med',np.sum(rmse_2d)/len(rmse_2d))
print('min',np.amin(rmse_3d),'med',np.sum(rmse_3d)/len(rmse_3d))
print('min',np.amin(rmse_4d),'med',np.sum(rmse_4d)/len(rmse_4d))
print('med',np.sum(rmse)/rmse.shape[0])
```
