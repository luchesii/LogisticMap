```python
t1000 = np.loadtxt("esn_big_minstable_r4_x0.1_d1000.csv", delimiter=",")
t5000 = np.loadtxt("esn_big_minstable_r4_x0.1_d5000.csv", delimiter=",")

cont=0
cont2=0
pts = []
for i in range (t5000.shape[0]):
    #print(t1000[i][3],t5000[i][3])
    if t5000[i][3]<=t1000[i][3]:
        cont=cont+1
        #print(t1000[i][3],t5000[i][3])
    if t5000[i][3]<=t1000[i][3] and t5000[i][1]==t1000[i][1] and t5000[i][2]==t1000[i][2]:
        cont2=cont2+1
    pts.append(cont/(i+1))

print(cont2)
t = np.arange(t5000.shape[0])
print(cont,cont/t5000.shape[0])
plt.plot(t, pts)
plt.show()

t500 = np.loadtxt("esn_big_minstable_r4_x0.1_d500.csv", delimiter=",")

cont=0
cont2=0
pts = []
for i in range (t500.shape[0]):
    #print(t1000[i][3],t5000[i][3])
    if t1000[i][3]<=t500[i][3]:
        cont=cont+1
        #print(t500[i][3],t1000[i][3])
    if t1000[i][3]<=t500[i][3] and t1000[i][1]==t500[i][1] and t1000[i][2]==t500[i][2]:
        cont2=cont+1
    pts.append(cont/(i+1))
    
print(cont2)

t = np.arange(t500.shape[0])
print(cont,cont/t500.shape[0])
plt.plot(t, pts)
plt.show()

cont=0
pts = []
for i in range (t5000.shape[0]):
    #print(t1000[i][3],t5000[i][3])
    if t5000[i][3]<=t500[i][3]:
        cont=cont+1
        #print(t500[i][3],t5000[i][3])
    pts.append(cont/(i+1))
    
t = np.arange(t5000.shape[0])
print(cont,cont/t5000.shape[0])
plt.plot(t, pts)
plt.show()
```
