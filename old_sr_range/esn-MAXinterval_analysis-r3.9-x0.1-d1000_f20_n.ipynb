```python
import numpy as np
from pyESN import ESN
from matplotlib import pyplot as plt
import random

def get_data(x0):
    r = 3.9
    x = []
    x.append(x0)
    for i in range(1,1100):
        x1=x0*r*(1-x0)
        x0=x1
        x.append(x1)
    x=np.asarray(x)
    return(x)

menor = np.loadtxt("esn_big_minstable_r3.9_x0.1_d1000.csv", delimiter=",")
list_of_x0 = np.loadtxt("esn_random_x0.1_r3.9.csv", delimiter=" ")


print(menor.shape)
```

```python
def get_array(erro_list):
    erro_array = np.asarray(erro_list).reshape(int(len(erro_list)/5),5)
    return erro_array

def get_array2(erro_list):
    erro_array = np.asarray(erro_list).reshape(int(len(erro_list)),20)
    return erro_array

    
def get_errors2(e0,e1,e2,e3,e4,e5,e6,e7,e8,e9,e10,e11,e12,e13,e14,e15,e16,e17,e18,e19):
    
    mae=[np.sum(np.abs(e0))/e0.shape[0],np.sum(np.abs(e1))/e1.shape[0],np.sum(np.abs(e2))/e2.shape[0],
       np.sum(np.abs(e3))/e3.shape[0],np.sum(np.abs(e4))/e4.shape[0],np.sum(np.abs(e5))/e5.shape[0],
       np.sum(np.abs(e6))/e6.shape[0],np.sum(np.abs(e7))/e7.shape[0],np.sum(np.abs(e8))/e8.shape[0],
       np.sum(np.abs(e9))/e9.shape[0],np.sum(np.abs(e10))/e10.shape[0],np.sum(np.abs(e11))/e11.shape[0],np.sum(np.abs(e12))/e12.shape[0],
       np.sum(np.abs(e13))/e13.shape[0],np.sum(np.abs(e14))/e14.shape[0],np.sum(np.abs(e15))/e15.shape[0],
       np.sum(np.abs(e16))/e16.shape[0],np.sum(np.abs(e17))/e17.shape[0],np.sum(np.abs(e18))/e18.shape[0],
       np.sum(np.abs(e19))/e19.shape[0]]
    
    mse=[np.sum(e0**2)/e0.shape[0],np.sum(e1**2)/e1.shape[0],np.sum(e2**2)/e2.shape[0],
       np.sum(e3**2)/e3.shape[0],np.sum(e4**2)/e4.shape[0],np.sum(e5**2)/e5.shape[0],
       np.sum(e6**2)/e6.shape[0],np.sum(e7**2)/e7.shape[0],np.sum(e8**2)/e8.shape[0],
       np.sum(e9**2)/e9.shape[0],
        np.sum(e10**2)/e10.shape[0],np.sum(e11**2)/e11.shape[0],np.sum(e12**2)/e12.shape[0],
       np.sum(e13**2)/e13.shape[0],np.sum(e14**2)/e14.shape[0],np.sum(e15**2)/e15.shape[0],
       np.sum(e16**2)/e16.shape[0],np.sum(e17**2)/e17.shape[0],np.sum(e18**2)/e18.shape[0],
       np.sum(e19**2)/e19.shape[0]]

    rmse=[np.sqrt(np.sum(e0**2)/e0.shape[0]),np.sqrt(np.sum(e1**2)/e1.shape[0]),np.sqrt(np.sum(e2**2)/e2.shape[0]),
       np.sqrt(np.sum(e3**2)/e3.shape[0]),np.sqrt(np.sum(e4**2)/e4.shape[0]),np.sqrt(np.sum(e5**2)/e5.shape[0]),
       np.sqrt(np.sum(e6**2)/e6.shape[0]),np.sqrt(np.sum(e7**2)/e7.shape[0]),np.sqrt(np.sum(e8**2)/e8.shape[0]),
       np.sqrt(np.sum(e9**2)/e9.shape[0]),
       np.sqrt(np.sum(e10**2)/e10.shape[0]),np.sqrt(np.sum(e11**2)/e11.shape[0]),np.sqrt(np.sum(e12**2)/e12.shape[0]),
       np.sqrt(np.sum(e13**2)/e13.shape[0]),np.sqrt(np.sum(e14**2)/e14.shape[0]),np.sqrt(np.sum(e15**2)/e15.shape[0]),
       np.sqrt(np.sum(e16**2)/e16.shape[0]),np.sqrt(np.sum(e17**2)/e17.shape[0]),np.sqrt(np.sum(e18**2)/e18.shape[0]),
       np.sqrt(np.sum(e19**2)/e19.shape[0])]
    
    return mae,mse,rmse
```

```python


future_error = []
future_error_t1 = []
future_error_t2 = []
future_error_t3 = []



for i in range(menor.shape[0]):
    data = get_data(list_of_x0[i])
    
    future_t = np.zeros((10))
    esn = ESN(n_inputs = 1,
        n_outputs = 1,
      n_reservoir = int(menor[i][1]),
      spectral_radius = menor[i][2],
      random_state=42)
    trainlen = 1000
    future = 20
    pred_training = esn.fit(np.ones(trainlen),data[0:trainlen])

    prediction = esn.predict(np.ones(future))
    error = np.sqrt(np.mean((prediction.flatten() - data[trainlen:trainlen+future])**2))
    print(i,'test error:',error)
   
    future_error.append(prediction.flatten()-data[trainlen:trainlen+future])
    if error>=1:
        future_error_t1.append(prediction.flatten()-data[trainlen:trainlen+future])
    elif error<1 and error>=0.1:
        future_error_t2.append(prediction.flatten()-data[trainlen:trainlen+future])
    elif error<0.01:
        future_error_t3.append(prediction.flatten()-data[trainlen:trainlen+future])

    
    
    
```
