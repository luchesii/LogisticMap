```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
table1 = np.loadtxt("mlp_x1_r3.997_minstable.csv", delimiter=",")
table2 = np.loadtxt("mlp_x2_r3.997_minstable.csv", delimiter=",")
table3 = np.loadtxt("mlp_x3_r3.997_minstable.csv", delimiter=",")
table4 = np.loadtxt("mlp_x4_r3.997_minstable.csv", delimiter=",")
print(table1.shape)
```

```python
for i in range(1,5):
    
    if i ==1: 
        N = table1[:,0]
        MAE = table1[:,1]
        MSE = table1[:,2]
        RMSE = table1[:,3]
    elif i ==2: 
        N = table2[:,0]
        MAE = table2[:,1]
        MSE = table2[:,2]
        RMSE = table2[:,3]
    elif i ==3: 
        N = table3[:,0]
        MAE = table3[:,1]
        MSE = table3[:,2]
        RMSE = table3[:,3]
    elif i ==4: 
        N = table4[:,0]
        MAE = table4[:,1]
        MSE = table4[:,2]
        RMSE = table4[:,3]
        
    print(N.shape,MAE.shape,MSE.shape)
    plt.plot(N,MSE)
    plt.title('MSE xn+%d' %(i))
    plt.ylabel('mse')
    plt.xlabel('Number of neurons')
    plt.show()
    
    plt.plot(N,RMSE)
    plt.title('RMSE xn+%d' %(i))
    plt.ylabel('rmse')
    plt.xlabel('Number of neurons')
    plt.show()
    
    plt.plot(N,MAE)
    plt.title('MAE xn+%d' %(i))
    plt.ylabel('mae')
    plt.xlabel('Number of neurons')
    plt.show()
    
    print('MSE min:',N[np.argmin(MSE)],MSE[np.argmin(MSE)])
    print('MAE min:',N[np.argmin(MAE)],MAE[np.argmin(MAE)])
    print('RMSE min:',N[np.argmin(RMSE)],RMSE[np.argmin(RMSE)])
```
