```python
import numpy as np
import keras
from keras.models import Sequential
import tensorflow as tf
import torch
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
dataset=np.loadtxt("lstm_xn+10_data1000_x0.1_r4_l10.csv", delimiter=",")

lenght=dataset.shape[1]-1
data=dataset[:,0:lenght].copy()
label=dataset[:,lenght:lenght+1].copy()

dtrain,dtest,ltrain,ltest=train_test_split(data,label,train_size=0.7,random_state=1)

print(lenght)
'''
dtrain=tf.convert_to_tensor(dtrain0)
dtest=tf.convert_to_tensor(dtest0)
ltest=tf.convert_to_tensor(ltest0)
ltrain=tf.convert_to_tensor(ltrain0)'''
```

```python
results = []

n = 5
while n<=1000:
    
    model = Sequential()
    model.add(keras.layers.LSTM(n, activation='relu', input_shape=(lenght, 1)))
    model.add(keras.layers.Dense(1)) 

    model.compile(optimizer=keras.optimizers.Adam(),loss='mean_squared_error')
    
    dtrain = dtrain.reshape((dtrain.shape[0],lenght,1))
    dtest = dtest.reshape((dtest.shape[0],lenght,1))
    history = model.fit(dtrain, ltrain, epochs=350,verbose=0)
    
    msetrain = model.evaluate(dtrain, ltrain)
    mse = model.evaluate(dtest, ltest)
    print(mse)
    
    line = np.empty((3))
    line = n, msetrain, mse
    
    results.append(line)
    
    print('n:',n,'\n mse:',mse)
    
    plt.plot(history.history['loss'])
    plt.title('Data:1000, lenght=10, n=%d' %n)
    plt.ylabel('mse')
    plt.xlabel('epoch')
    plt.show()
    
    
    if n < 50:
        n = n+5
    elif n < 100:
        n = n+10
    elif n < 1000:
        n = n+100
    else:
        n = n+1000
        
print(results)
            


    
```
