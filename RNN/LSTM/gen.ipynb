```python
import numpy as np
import random

r = 4
x0=0.1

n=1000 #n√∫mero de dados no dataset

lenght=10 #tamanho da sequencia

file = open('lstm_xn+10_data1000_x0.1_r4_l10.csv','w+')
```

```python
for i in range(200):
    x1=x0*r*(1-x0)
    x0=x1

x=[]
for i in range(n*10):
    x1=x0*r*(1-x0)
    x0=x1
    x.append(x1)
    
x=np.asarray(x)

ind=random.sample(range(x.shape[0]-lenght**2-9), n)

print(ind)


for i in range(n):
    for j in range(lenght):
        file.write('{}'.format(x[ind[i]+j]))
        file.write(',')
    file.write('{}'.format(x[ind[i]+lenght+9]))
    file.write('\n')
    
file.close()
```
