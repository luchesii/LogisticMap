```python
import numpy as np
import matplotlib.pyplot as plt 

np.set_printoptions(threshold=np.nan)

x0=0.1
r=3.8

n_new=20 #número de pontos novos a serem previstos

for i in range(200): #eliminar o transiente
    x1=x0*r*(1-x0)
    x0=x1
    
lenght=100 #tamanho da série temporal

x_timeser=[]
for i in range(lenght):
    x1=x0*r*(1-x0)
    x0=x1
    x_timeser.append(x1)

x_timeser=np.asarray(x_timeser) #série temporal

x_timeser_r=x_timeser.copy() #cópia da série original que nunca será modificada
x_timeser_o=x_timeser.copy() #cópia da série, no final serão adicionados os novos pontos reconstrúidos pelo método

t=np.arange(0,lenght,1)

plt.plot(t,x_timeser,'b')
plt.title('time series')
plt.show()
```

```python
def lyapunov(x0,r): #função que calcula lyapunov
    
    for i in range(200): 
        x1=x0*r*(1-x0)
        x0=x1
    
    lyp=0
    for i in range(1000):
        x1=x0*r*(1-x0)
        x0=x1
        lyp=lyp+np.log(abs(r-2*r*x1))
    lyp=lyp/10000
    
    return lyp
```

```python
D=2

coor_emb=np.empty((lenght-D+1,D)) #embedding na dimensão D

for i in range (coor_emb.shape[0]):
    for j in range (D):
        coor_emb[i][j]=x_timeser[i+j] #pegando as coordenadas no espaço de dimensão D

plt.plot(coor_emb[:,0],coor_emb[:,1],'ro')
plt.title('Embedding 2D')
plt.show()


```

```python
for l in range(lenght,lenght+n_new):
    
    m_dist=np.zeros((l-D)) #guardar distâncias entre o último vetor e todos os outros.
    
    for i in range (l-D):
        m_dist[i]=np.linalg.norm(coor_emb[l-D]-coor_emb[i])
        
    for i in range(l-D):
        print(i,coor_emb[i],m_dist[i])
        
    print(l, coor_emb[l-D])

    ind_min=np.argmin(m_dist) #acha qual vetor está mais próximo do último
    dist_0=np.min(m_dist) #dá distância do vetor mais próximo ao último
    
    print(ind_min,dist_0)
    
    lyap=lyapunov(x0,r)
    
    dist=[]
    for k in range (15):
        a=dist_0*2**(k*lyap)
        dist.append(a)
    
    dist=np.asarray(dist)
    
    dist_1=dist[0] #calcula distância que deve existir enre o vetor ind_min+1 e l-(D+1)+1
    
    #bhaskara
    b=-2*coor_emb[ind_min+1][D-1]
    c_aux=0
    for i in range (D-1):
        c_aux=c_aux+(coor_emb[ind_min+1][i]-coor_emb[l-D][i+1])**2
        
    c=coor_emb[ind_min+1][D-1]**2 -dist_1**2 +c_aux
    
    new_row=np.empty((1,D)) #novo vetor
    
    for i in range(D-1):
        new_row[0][i]=coor_emb[l-D][i+1] #pelo método, é fácil ver que as primeiras coordenadas do novo vetor são essas
        
    new_row[0][D-1]=(-b+np.sqrt(b**2-4*c))/2 #novo valor (último) do vetor calculado
    
    print(new_row) #novo vetor 
    
    coor_emb = np.vstack([coor_emb, new_row]) #adicionando novo vetor na matriz coor_emb
    print(coor_emb[l-D+1])
    
    #print(np.linalg.norm(coor_emb[l-D+1]-coor_emb[ind_min+1]),dist_1)
    
    x1=x_timeser[l-1]*r*(1-x_timeser[l-1]) #novo valor da serie temporal esperado
    print(x1,coor_emb[l-D+1,D-1])
    
    
    x_timeser=np.hstack([x_timeser,x1])
    x_timeser_r=np.hstack([x_timeser_r, coor_emb[l-D+1,D-1]])

    
error=x_timeser_r-x_timeser
mse_f=error**2
mae_f=abs(error)
res_f=error/x_timeser

mae_f=mae_f[lenght:]
mse_f=mse_f[lenght:]
res_f=res_f[lenght:]
error=error[lenght:]

t=np.arange(lenght,lenght+n_new,1)

plt.plot(t,error,'g')
plt.title('error')
plt.show()

plt.plot(t,mse_f,'b')
plt.title('mse')
plt.show()


mae=sum(mae_f)/mae_f.shape[0]
mse=sum(mse_f)/mse_f.shape[0]
res=sum(res_f)

print('mae:',mae,'\n mse:',mse,'\n res:',res)
```
